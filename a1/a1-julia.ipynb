## Assignment 1 - Julia solution

```julia
using CSV, DataFrames
```

We have imported CSV and DataFrames to work with the data!

```julia
df = CSV.File("data.csv") |> DataFrame
```

```julia
function filterFunc(row)
  
  if(row.Code == "" || occursin("OWID_", row.Code))
    return false
  end
  
  if(row.Year >= 2020 || row.Year < 2010)
    return false
  end
  print(row)
  return true
end
filter(filterFunc, df)
```
